## Real World Examples 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing data

```python
# The data is imported from the file called "gas_prices.csv"
```

```python
gp = pd.read_csv("gas_prices.csv")
gp.head()
```

```python
gp.info()
```

```python
## A simple line graph for a particular country in the data set

plt.plot(gp['Year'], gp['Australia'])
plt.show()
```

```python
## Now let's see more than one country in one graph
## On the graph at the previous cell and the one I ran just underneath, I realized the x axis should be broader to increase readability
## Therefore, after I run the code for the first time, I changed the outer dimensions of the line chart (The first line of code in this cell)

plt.figure(figsize=(12,7))
plt.plot(gp['Year'], gp['Australia'],'b.-', label = 'Australia')
plt.plot(gp['Year'], gp['Japan'],'g.-', label = 'Japan')
plt.plot(gp['Year'], gp['South Korea'], 'y.-', label = 'South Korea')


plt.title("Gas Prices in USD over the years")
plt.xlabel("Years")
plt.ylabel("Gas Prices in USD")
plt.xticks(gp["Year"])
plt.legend(labelcolor = 'black', loc = 'lower right')

plt.show()
```

```python
## Now let's see all the countries without manually writing them one by one

plt.figure(figsize=(15,10))

for c in gp.drop(['Year'], axis=1):
    plt.plot(gp["Year"], gp[c], '.-', label = c)
    

plt.title("Gas Prices in USD over the years",  fontsize = 30, fontweight = "bold")
plt.xlabel("Years", fontsize = 16)
plt.ylabel("Gas Prices in USD", fontsize = 16)
plt.xticks(gp["Year"])
plt.legend(labelcolor = 'black', loc = 'upper left')


plt.savefig("Gas Prices in USD over the years.png", dpi =200)
plt.show()    
```
