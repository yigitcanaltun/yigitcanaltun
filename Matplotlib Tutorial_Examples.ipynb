```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1,3,5,7,9]
y = [2,6,10,14,18]

#plt.plot(x,y, color='green', marker='o', linestyle='-', 
     #linewidth=2.5, markersize=12, markeredgecolor = 'black', markerfacecolor = 'black', label = 'random numbers')

## shorthand     
    
plt.plot(x,y, 'b*--', label = '2 times x') ## fmt = '[color][marker][line]'

plt.title('Example', fontdict = {'fontname': 'Arial', 'fontsize': 30})
plt.xlabel('X title', fontdict = {'fontname': 'Arial', 'fontsize': 17})
plt.ylabel('Y title', fontdict = {'fontname': 'Arial', 'fontsize': 17})

plt.xticks([0,2,4,6,8,10])
plt.yticks([0,4,8,12,16,20]) # the ticks can mismatch in terms of the dimension

plt.legend()

plt.show()
```

### Showing two different functions in one graph with different colors, shapes, markers etc.

```python
plt.figure(figsize=(5,3), dpi=200)
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**3, 'r^-', label = 'X cube')
plt.plot(x2[5:], x2[5:]**3, 'r^--', label = 'X cube')

plt.plot(x,y, 'b*--', label = '2 times x')
plt.xticks([0,2,4,6,8,10])
plt.yticks([0,20,40,60,80,100])
plt.legend()

plt.savefig('mygraph.png', dpi =300) # dpi can be different than the one above

plt.show()
```
